```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python

x=np.linspace(0,10,100) # 100 points from 0 to 10 
y=np.sin(x)
plt.figure(figsize=(8,5)) #width,height in inches
plt.plot(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.title("sin-graph")
plt.show()
```

```python
x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)
y3=np.sin(x)*np.cos(x)
plt.figure(figsize=(10,6))
plt.plot(x,y1,label='sin(x)')
plt.plot(x,y2,label='cos(x)')
plt.plot(x,y3,label='sin(x)*cos(x)')
plt.xlabel("x-axis")
plt.ylabel("y-label")
plt.title("multiple graph")
plt.legend() # that legend with color 
plt.grid(True,alpha=0.3) # for transperency 
plt.show()
```

```python
x=np.linspace(0,10,20)
plt.figure(figsize=(10,6))
plt.plot(x,x,'b--',label='solid line',linewidth=2)
plt.plot(x,x+1,'r--',label="dashed line",linewidth=2)
plt.plot(x,x+2,'g-.',label="dash dot line",linewidth=2)
plt.plot(x,x+3,'m:',label='dotted line',linewidth=2)
plt.plot(x,x+4,'co-',label="circles",markersize=8)
plt.plot(x,x+5,"s-",label="sqaures",markersize=6)
plt.plot(x,x+6,"^-",label="traingles",markersize=6)

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("so beautiful lines")
plt.legend()
plt.grid(True,alpha=0.3)
plt.show()
```

Generate sample data

```python
# 1. Initialize the modern random number generator with a seed
rng = np.random.default_rng(seed=42)

# 2. Generate sample data using standard_normal() instead of randn()
x = rng.standard_normal(100)
y = 2 * x + rng.standard_normal(100) * 0.5  # Linear relationship with noise
```

```python

plt.figure(figsize=(5,4))
plt.scatter(x,y,alpha=0.6)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("scater plot")
plt.grid(True,alpha=0.3)
plt.show()
```

```python
rng=np.random.default_rng(seed=3)
x=rng.random(100)
y=rng.random(100)
colors=rng.random(100)
sizes=1000*rng.random(100)

plt.figure(figsize=(8,5))
plt.scatter(x,y,c=colors,s=sizes,alpha=0.6,cmap='plasma')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("scatter plot with colore size variation")
plt.grid(True,alpha=0.3)
plt.show()
```

BAR CHART

```python
cat=["A","B","C","D","E"]
val=[15,83,38,78,89]

plt.figure(figsize=(7,5))
plt.bar(cat,val,color="hotpink",edgecolor="black",linewidth=1)  # bar to barh for horizonatl
plt.xlabel("categories")
plt.ylabel("values")
plt.title("simple bar chart")
plt.grid(True,alpha=0.4,axis="y")  # x to y for horizonatal
plt.show()
```

double bar chart

```python
cat=["q1","q2","q3","q4"]
pr_a=[20,30,40,50]
pr_b=[22,45,67,83]
pr_c=[34,56,72,33]

x=np.arange(len(cat))
width=0.2

plt.figure(figsize=(10,5))

plt.bar(x-width,pr_a,width,color="lightgreen",label="Product A")
plt.bar(x,pr_b,width,color="orange",label="product b")
plt.bar(x+width,pr_c,width,color="lightblue",label="product c")

plt.scatter(x - width, pr_a, color="darkgreen", zorder=3, label="A Points")
plt.scatter(x, pr_b, color="darkorange", zorder=3, label="B Points")
plt.scatter(x + width, pr_c, color="darkblue", zorder=3, label="C Points")

plt.xlabel("quarter")
plt.ylabel("sales")
plt.title("monthly sales you know well")
plt.xticks(x,cat)
plt.legend()
plt.grid(True,alpha=0.3,axis="y")
plt.show()
```
